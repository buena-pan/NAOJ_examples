```python
import numpy as np
import matplotlib.pyplot as plt

def compute_fringe_pattern(wavelength, baseline, theta_range):
    # Calculate the path difference due to the baseline and angle
    delta_path = baseline * np.sin(theta_range)
    
    # Calculate the phase difference
    delta_phi = 2 * np.pi * delta_path / wavelength
    
    # Intensity using a simple two-telescope interference model
    # I = I1 + I2 + 2*sqrt(I1*I2)*cos(delta_phi)
    # Assuming I1 = I2 = 1 for simplicity
    intensity = 2 * (1 + np.cos(delta_phi))
    
    return intensity

# Define parameters
wavelength = 1e-1  # 10 cm, radio wave
baseline = 5.0  # 5 meter
theta_range = np.linspace(-2*np.pi, 2*np.pi, 5000)

intensity = compute_fringe_pattern(wavelength, baseline, theta_range)

# Plot
plt.rcParams["figure.figsize"] = (20,3)
plt.plot(theta_range, intensity)
plt.title('Interference Fringe Pattern')
plt.xlabel('Angle (radians)')
plt.ylabel('Intensity')
plt.grid(True)
plt.show()
```
