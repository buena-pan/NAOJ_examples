```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display
```

```python

# Parameters
f = 1  # Frequency of sine wave
t = np.linspace(0, 4 * np.pi, 1000)  # Time values
original_sine = np.sin(f * t)  # Original sine wave
noise = np.random.normal(0, 0.5, t.shape)  # Random noise
noisy_sine = original_sine + noise  # Noisy sine wave

# Reducing noise using moving average (integration)
window_sizes = [5, 50, 150 ,500]  # Different window sizes for averaging
averaged_signals = []

for window in window_sizes:
    averaged_signal = np.convolve(noisy_sine, np.ones(window)/window, mode='same')
    averaged_signals.append(averaged_signal)

# Plotting
plt.figure(figsize=(12, 8))

plt.subplot(4, 2, 1)
plt.plot(t, original_sine, label="Original Sine", color='blue')
plt.title("Original Sine Wave")
plt.legend()

plt.subplot(4, 2, 2)
plt.plot(t, noisy_sine, label="Noisy Sine", color='red')
plt.title("Noisy Sine Wave")
plt.legend()

for idx, window in enumerate(window_sizes):
    plt.subplot(4, 2, idx + 3)
    plt.plot(t, averaged_signals[idx], label=f"Window Size {window}")
    plt.plot(t, original_sine, label="Original Sine", linestyle="--", color='blue')
    plt.legend()
    plt.title(f"Filtered Signal with Window Size {window}")

plt.tight_layout()
plt.show()
```
