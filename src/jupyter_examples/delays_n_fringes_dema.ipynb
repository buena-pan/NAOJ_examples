```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_intensity(I1, I2, wavelength, d):
    """Calculate combined intensity based on path difference d."""
    delta_phi = (2 * np.pi / wavelength) * d
    return I1 + I2 + 2 * np.sqrt(I1 * I2) * np.cos(delta_phi)

def simulate_interference(I1, I2, wavelengths, ds):
    """Simulate interference for given intensities, wavelengths and path differences."""
    intensities = np.zeros_like(ds)
    for wavelength in wavelengths:
        intensities += calculate_intensity(I1, I2, wavelength, ds)
    return intensities

def main():
    # Equal intensities for the two beams
    I1 = 1  
    I2 = 1  

    # Wavelengths (considering a narrow band around 650 nm for simplicity)
    wavelengths = np.linspace(640e-9, 660e-9, 5)  

    # Path differences ranging from -2 wavelengths of 650nm to +2 wavelengths
    ds = np.linspace(-2*650e-9, 2*650e-9, 1000)
    intensities = simulate_interference(I1, I2, wavelengths, ds)

    plt.plot(ds, intensities)
    plt.title("Intensity vs Path Difference")
    plt.xlabel("Path Difference (m)")
    plt.ylabel("Intensity")
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    main()
```
